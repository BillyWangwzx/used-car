We have 83 variables. Price for sale is the variable we want to predict. However, The Price for sale is greatly affected by the price of a new car. And the price for sale is bounded between 0 and new price. Instead of predicting the price, we could use de depreciation rate($\frac{price\space for\space sale}{new\space car\space price}$) as the response.

Fisrt we build a baseline model and choose some variables that effect the depreciate rate in common sense, like used time, miles, the type of transmission, new car price(expensive car would depreciate faster than cheap car to some extent)

```python
baseline_model_data = pd.DataFrame({
    "depreciation_rate" : data['卖家报价(万)']/data['新车指导价(万)'],
    "time" : data['上牌时间'].apply(lambda x: 2017-int(x.split('-')[0])+(6-int(x.split('-')[1]))/12.0),
    "mile" : data['公里数'].apply(lambda x: float(x[:-3])),
    "transmission" : data['变速箱1'].apply(lambda x: 1 if x=="自动" else 1),
    "newprice" : data['新车指导价(万)'],
    "price" : data['卖家报价(万)']
})
```

```python
baseline_model_data
```

we need to choose a function. As the depreciation rate is bounded between 0 and 1, we could use the logistic function for this circumstance. $depreciation\_rate = \frac{exp(\beta_0+\beta_1mile+\beta_2newprice+\beta_3time+\beta_4transmission+\epsilon)}{1+exp(\beta_0+\beta_1mile+\beta_2newprice+\beta_3time+\beta_4transmission+\epsilon)}$

Apply a transformation to this function, we can get.

$log(\frac{1}{1-depreciation\_rate})=\beta_0+\beta_1mile+\beta_2newprice+\beta_3time+\beta_4transmission+\epsilon$

```python
baseline_model_data = baseline_model_data.assign(y=np.log(1/(1-baseline_model_data.depreciation_rate)))
```

```python
from sklearn import linear_model
```

```python
baseline = linear_model.LinearRegression()
```

```python
baseline.fit(baseline_model_data.drop(['depreciation_rate','y'],axis = 1),baseline_model_data['y'])
```

```python
baseline.score(baseline_model_data.drop(['depreciation_rate','y'],axis = 1),baseline_model_data['y'])
```
